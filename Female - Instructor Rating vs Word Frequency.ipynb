```python
import scattertext as st
import spacy
import pandas as pd
```

```python
df = pd.read_csv('STEM_Female_Sample.csv')
```

```python
# Pt. (1): Clean datafile and establish axes 
# Convert Scores from string to float

df['Score'] = df['Score'].astype(float)
```

```python
# Establish binary categories (x- and y-axis)
# Convert Scores from floats to one of two categories

df['Score'] = df['Score'].apply(lambda x: 'High Rating' if x > 4 else 'Low Rating')
```

```python
# Pt. (2): Apply NLP to datafile

nlp = spacy.load('en')
```

```python
# Establish stop words

from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
nlp.Defaults.stop_words |= stopWords

# Add textfile to add custom stop words
with open('stopwords.txt', 'r') as f:
    str_f = f.read()
    stopwords_file = set(str_f.split('\n'))
nlp.Defaults.stop_words |= stopwords_file
```

```python
# Builds corpus ('Score' v. Word Frequency in 'Comment'), and removes stopwords

corpus = st.CorpusFromPandas(df, 
                            category_col='Score',
                            text_col='Comment',
                            nlp=nlp).build().remove_terms(nlp.Defaults.stop_words, ignore_absences=True)
```

```python
# Pt. (3): Visualize term associations with Scattertext

# Builds html file
html = st.produce_scattertext_explorer(corpus,
                                       category='Low Rating',
                                       category_name='Low Rating',
                                       not_category_name='High Rating',
                                       width_in_pixels=1000,
                                       height_in_pixels=700)
```

```python
# Saves html file
open("Female-Instructor-Rating-Vs-Word-Frequency.html", 'wb').write(html.encode('utf-8'))
```
